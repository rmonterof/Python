##Barras


```python
x = np.array(["Producto1", "Producto2", "Producto3", "Producto4"])
y = np.array([5.0, 9.99, 7.99, 3.21])
plt.bar(x,y)
```

```python
x = np.array(["Producto1", "Producto2", "Producto3", "Producto4"])
y = np.array([5.0, 9.99, 7.99, 3.21])
plt.barh(x,y)
```

```python
x = np.array(["Producto1", "Producto2", "Producto3", "Producto4"])
y = np.array([5.0, 9.99, 7.99, 3.21])
plt.barh(x,y, color = "purple")
```

```python
x = np.array(["Producto1", "Producto2", "Producto3", "Producto4"])
y = np.array([5.0, 9.99, 7.99, 3.21])

index = np.arange(len(x))
index
```

```python
x = np.array(["Producto1", "Producto2", "Producto3", "Producto4"])
y = np.array([5.0, 9.99, 7.99, 3.21])

index = np.arange(len(x))
plt.bar(index, y)
plt.xticks(index, x)
plt.title("Precios")
```

```python
df = pd.Series({
    "Producto" :["Producto1", "Producto2", "Producto3", "Producto4"],
    "Precios" : [5.0, 9.99, 7.99, 3.21]
})

plt.bar(df["Producto"], df["Precios"])
```

```python
etiquetas = ['Enero', 'Febrero', 'Marzo']
ingresos = [1500.0, 2000.0, 1700.0]
gastos= [1200.0, 1000.0, 1800.0]

index = np.arange(len(etiquetas))
plt.bar(index, ingresos, color= "green")
plt.bar(index, gastos, color= "red")
plt.xticks(index, etiquetas)
```
